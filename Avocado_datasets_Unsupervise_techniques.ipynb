# **CLUSTERING**

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=106,random_state=40)
model.fit(x)
```

```python
x
```

```python
model.cluster_centers_
```

```python
sns.scatterplot(data=x,y=4225, x='Total Volume')
centre_x, centre_y = model.cluster_centers_[:,4], model.cluster_centers_[:, 2]
plt.plot(centre_x, centre_y,'xb')
```

```python
model.inertia_
```

```python
model=KMeans(n_clusters=10,random_state=15)
model.fit(x)
```

```python
model.cluster_centers_
```

```python
sns.scatterplot(data=x,x='Total Bags',y='AveragePrice')
x_centre , y_centre =model.cluster_centers_[:, 1],model.cluster_centers_[:, 6]
plt.plot(x_centre,y_centre,'xb')
```

```python
model.inertia_
```

```python
model.predict(x)
```

```python
options =range(2,11)
inertias = []

for n_clusters in options:
  model = KMeans(n_clusters,random_state=0).fit(x)
  inertias.append(model.inertia_)

plt.title('No. of clusters VS Inertia')
plt.plot(options,inertias,'-o')
plt.xlabel('No. of clusters')
plt.ylabel('Inertia')
```

```python
from sklearn.cluster import DBSCAN
model=DBSCAN(eps= 1.5,min_samples=8)
model.fit(x)
```

```python
model.core_sample_indices_
```

```python
labels =model.labels_
labels
```
