# Pie Charts

Proportion of Total Houses Sold by Area

```python
house_sold_ar = df.groupby('area')['houses_sold'].sum().dropna()
fig = px.pie(house_sold_ar, values = 'houses_sold', names = house_sold_ar.index , title = 'Proportion of total house by area')
fig.show()
```

2. Proportion of Areas with and Without Borough Status

```python
borough_counts =df['borough_flag'].value_counts()
fig=px.pie(borough_counts , values = borough_counts.values , names =['Borough','Not Borough'] , title ='Proportion of borough vs non borough')
fig.show()
```

Distribution of **Recorded Crimes** by Area

```python
area_crime = df.groupby('area')['no_of_crimes'].sum().dropna()
fig = px.pie(area_crime , values ='no_of_crimes',names = area_crime.index, title = 'distribution of Record Cirme by Area' )
fig.show()
```
