```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style('darkgrid')
%matplotlib inline
```

```python
iris_df =sns.load_dataset('iris')
iris_df
```

```python
sns.scatterplot(data = iris_df,x= 'sepal_length',y='petal_length',hue = 'species')
```

**We'll attempt to cluster observations using numeric columns in the data**

```python
numeric_col = ['sepal_length','sepal_width','petal_length','petal_width']
x=iris_df[numeric_col]
x
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters= 3,random_state= 35)
model.fit(x)
```

```python
pred = model.predict(x)
pred
```

```python
sns.scatterplot(data=x,x='sepal_length',y='petal_length',hue=pred);
centre_x, centre_y = model.cluster_centers_[:, 0],model.cluster_centers_[:, 2]
plt.plot(centre_x, centre_y, 'xb')
```

```python
model.inertia_
```

```python
model = KMeans(n_clusters=6,random_state=45).fit(x)
model
```

```python
pred =model.predict(x)
pred
```

```python
sns.scatterplot(data=x,x='sepal_length',y='petal_length',hue=pred);
centre_x, centre_y = model.cluster_centers_[:, 0],model.cluster_centers_[:, 2]
plt.plot(centre_x, centre_y, 'xb')
```

```python
model.inertia_
```

```python
model =KMeans(n_clusters =10,random_state = 65).fit(x)
model
```

```python
pred = model.predict(x)
pred
```

```python
sns.scatterplot(data=x,x='petal_length',y='petal_width',hue=pred);
x_centre, y_centre = model.cluster_centers_[:, 2],model.cluster_centers_[:, 3]
plt.plot(x_centre,y_centre,'xb')
```

```python
model.inertia_
```

```python
options = range(2,11)
inertias = []

for n_clusters in options:
    model = KMeans(n_clusters, random_state=42).fit(x)
    inertias.append(model.inertia_)

plt.title("No. of clusters vs. Inertia")
plt.plot(options, inertias, '-o')
plt.xlabel('No. of clusters (K)')
plt.ylabel('Inertia');
```

**Density-based spatial clustering of applications with noise (DBSCAN)**

```python
from sklearn.cluster import DBSCAN
```

```python
?DBSCAN
```

```python
model = DBSCAN(eps= 1.3, min_samples = 6).fit(x)
model
```

**In DBSCAN, there's no prediction step. It directly assigns labels to all the inputs.**



```python
labels = model.labels_
labels
```

```python
sns.scatterplot(data=x,x='sepal_length',y='petal_length',hue = labels);
```

```python
model=DBSCAN(eps=0.6,min_samples = 8).fit(x)
model
```

```python
labels = model.labels_
labels
```

```python
sns.scatterplot(data=x,x='sepal_length',y='petal_length',hue=labels);
```

```python
model=DBSCAN(eps=1.3,min_samples = 50).fit(x)
labels=model.labels_
labels
```

```python
sns.scatterplot(data=x,x='sepal_length',y='petal_length',hue=labels);
```
