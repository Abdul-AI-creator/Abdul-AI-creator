```python
df=pd.read_csv('/content/us-software-engineer-jobs/us-software-engineer-jobs-zenrows.csv')
df
```

```python
conn = sqlite3.connect('usa_company.db')
df.to_sql('engineers_data',con=conn,if_exists='replace',index=False)
cur=conn.cursor()
cur.execute('select count(*) from engineers_data')
print(f'no of rows in database is: {cur.fetchone()[0]}')
```

**Queries**

```python
q1 = '''
select distinct(company),max(salary) from engineers_data
'''
res=pd.read_sql_query(q1,conn)
res
```

top **10 highest paying** companies

```python
q = '''
select distinct(company) from engineers_data order by salary desc limit 10
'''
res=pd.read_sql_query(q,conn)
res
```

**[ indeed_apply_enabled company](https://)**

```python
q1 = '''
select title,company,salary,indeed_apply_enabled from engineers_data where indeed_apply_enabled == 1 group by salary
'''
res=pd.read_sql_query(q1,conn)
res
```

**distinct location where company present**

```python
q1 = '''
select distinct(location),company,salary from engineers_data
'''
res=pd.read_sql_query(q1,conn)
res
```

**COMPANIES WORK REMOTELY**

```python
q1 = '''
select company,title,location,remote_work_model,job_link from engineers_data where remote_work_model = 'REMOTE_ALWAYS'
'''
res=pd.read_sql_query(q1,conn)
res
```

**HIRING COMPANIES**

```python
q1 = '''
select company,title,new_job,remote_work_model,job_link from engineers_data where new_job = 1
'''
res=pd.read_sql_query(q1,conn)
res
```

**max reviews**

```python
q1 = '''
select company,review_count,urgently_hiring from engineers_data where review_count > 100 and urgently_hiring = 1
'''
res=pd.read_sql_query(q1,conn)
res
```

**$45,000 - $55,000 a year salary**
