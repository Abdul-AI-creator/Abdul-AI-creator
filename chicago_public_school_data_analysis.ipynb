```python
import sqlite3
import numpy as np
import sqlalchemy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("/content/Chicago_Public_Schools_-_Progress_Report_Cards__2011-2012__20240727.csv")
df
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.head()
```

```python
df.rename(columns={'elementary,_middle,_or_high_school':'e_m_hs'})
df
```

```python
conn = sqlite3.connect('chicago_institute.db')
df.to_sql('hs_school',con=conn,if_exists='replace',index=False)
cur = conn.cursor()
cur.execute('select * from hs_school')
print(f"the company records shown below: {cur.fetchone()[0]} ")
```

```python
df.info()
```

**creating view useful data**

```python
q='''
create view halfdata as select school_id,name_of_school,street_address,city,state,family_involvement_icon,safety_score from hs_school
'''
cur.execute(q)
conn.commit()
```

```python
q='''
select * from halfdata
'''
res =pd.read_sql_query(q,conn)
res
```

**more safe schools**

```python
q='''
select name_of_school,city from  halfdata where safety_score > (select min(safety_score) from halfdata) order by safety_score
'''
r =pd.read_sql_query(q,conn)
r
```

**less safe school**

```python
q='''
select name_of_school,city from  halfdata where safety_score < (select max(safety_score) from halfdata) order by safety_score
'''
r =pd.read_sql_query(q,conn)
r
```
